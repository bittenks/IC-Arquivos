```python
import pandas as pd
import numpy as np
```

```python
url1 = ('https://github.com/bittenks/IC-repository-tables/raw/main/corn2015_2020.xlsx')
corn= pd.read_excel(url1)

url2 = ('https://github.com/bittenks/IC-repository-tables/raw/main/soybean2015_2020.xlsx')
soybean = pd.read_excel(url2)

url3 = ('https://github.com/bittenks/IC-repository-tables/raw/main/cotton2015_2020.xlsx')
cotton = pd.read_excel(url3)

url4 = ('https://github.com/bittenks/IC-repository-tables/raw/main/wheat2018_2020.xlsx')
wheat = pd.read_excel(url4)
```

```python
corn
```

```python
corn = corn.info()
```

```python
cotton = cotton.info()
```

```python
soybean = soybean.info()
```

```python
wheat = wheat.info()
```

```python
test_corn_soy = pd.merge_ordered(corn, soybean, fill_method="ffill", left_by="Date")
test_corn_soy
```

```python
print(cotton)
```

```python
test_cot_wheat = pd.merge_ordered(cotton, wheat, fill_method="ffill", left_by="Date")
test_cot_wheat
```

```python
all_merge = pd.merge_ordered( test_corn_soy, test_cot_wheat, fill_method="ffill", left_by="Date")
```

```python
all_merge
```

```python
all_merge.to_csv('all_merge.csv')
```
